```python
# HASHIR, 12525400
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV
df = pd.read_csv(r"/content/jsw FINAL.REAL11.csv")
# Remove extra spaces from column names
df.columns = df.columns.str.strip()
# Convert Year to string
df['Year'] = df['Year'].astype(str)
# Clean Total Assets column
df['Total Assets'] = (
    df['Total Assets']
    .astype(str)
    .str.replace(',', '', regex=False)
    .replace('nan', None)
)
# Convert to numeric safely
df['Total Assets'] = pd.to_numeric(df['Total Assets'], errors='coerce')
# Remove empty rows
df = df.dropna(subset=['Total Assets'])
# Sort by Year
df.sort_values('Year', inplace=True)
# Plot graph
plt.plot(df['Year'],
         df['Total Assets'],
         marker='o',
         color='teal')
plt.title('Tata Steel - Total Assets')
plt.xlabel('Year')
plt.ylabel('Total Assets (₹ Cr)')
plt.xticks(rotation=45)
plt.show()
```
